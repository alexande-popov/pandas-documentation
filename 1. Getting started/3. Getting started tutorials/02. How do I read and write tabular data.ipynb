    В этом руководстве используется набор данных Titanic, сохраненный в формате CSV. Данные состоят из следующих столбцов данных:
    
    * PassengerId: идентификатор каждого пассажира.
    * Survived: указывает, выжил ли пассажир. `0` за да и `1` за нет.

    * Cabin: Номер кабины пассажира.

# Как читать и записывать табличные данные?

![02_io_readwrite](https://pandas.pydata.org/docs/_images/02_io_readwrite.svg)

___

⸮ Я хочу проанализировать данные о пассажирах Титаника, доступные в виде CSV-файла.

```python
titanic = pd.read_csv("data/titanic.csv")
```

pandas предоставляет функцию `read_csv()` для чтения данных, хранящихся в виде файла csv, в `DataFrame`. pandas поддерживает множество различных форматов файлов или источников данных из коробки (csv, excel, sql, json, parquet, …), каждый из них имеет префикс `read_*`.

Всегда проверяйте данные после чтения данных. При отображении `DataFrame` по умолчанию будут отображаться первые и последние 5 строк:

```python
titanic
```

___

⸮ Я хочу увидеть первые 8 строк `DataFrame`.

```python
titanic.head(8)
```

*Примечание:* Вместо этого вас интересуют последние N строк? pandas также предоставляет метод `tail()`. Например, `titanic.tail(10)` вернет последние 10 строк DataFrame.

```python
titanic.tail(10)
```

Проверить, как pandas интерпретирует каждый из типов данных столбца, можно, запросив атрибут pandas `dtypes`.

```python
titanic.dtypes
```

Для каждого столбца указывается используемый тип данных. Типы данных в этом `DataFrame` — целые числа (`int64`), числа с плавающей запятой (`float64`) и строки (`object`).

*Замечание:* При запросе `dtypes` скобки не используются! `dtypes` — это атрибут `DataFrame` и `Series`, а не метод.

___

⸮ Мой коллега запросил данные Титаника в виде электронной таблицы.

```python
titanic.to_excel("data/titanic.xlsx", sheet_name="passengers", index=False)
```

В то время как функции `read_*` используются для чтения данных в pandas, методы `to_*` используются для хранения данных. Метод `to_excel()` сохраняет данные в виде файла Excel. В приведенном здесь примере `sheet_name` называется `passengers`, а не листом по умолчанию `Sheet1`. При установке `index=False` метки индексов строк не сохраняются в электронной таблице.

Эквивалентная функция чтения `read_excel()` перезагрузит данные в `DataFrame`:

```python
titanic = pd.read_excel("data/titanic.xlsx", sheet_name="passengers")
titanic.head()
```

___

⸮ Меня интересует техническое резюме `DataFrame`

```python
titanic.info()
```

Метод `info()` предоставляет техническую информацию о `DataFrame`, поэтому давайте объясним вывод более подробно:

* Это действительно `DataFrame`.
* Имеется 891 запись, т.е. 891 строка.
* Каждая строка имеет метку строки (она же `index`) со значениями в диапазоне от 0 до 890.
* В таблице 12 столбцов. Большинство столбцов имеют значение для каждой из строк (все значения 891 `non-null`). В некоторых столбцах отсутствуют значения и менее 891 ненулевого значения.
* Столбцы `Name`, `Sex`, `Cabin` и `Embarked` состоят из текстовых данных (строки, также известные как `object`). Другие столбцы представляют собой числовые данные, причем некоторые из них — целые числа (так называемые `integer`), а другие — действительные числа (так называемые `float`).
* Типы данных (символы, целые числа и т. д.) в разных столбцах суммируются путем перечисления `dtypes`.
* Также указывается приблизительный объем оперативной памяти, используемой для хранения `DataFrame`.
